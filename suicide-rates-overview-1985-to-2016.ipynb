## Create Pivot Table <a name="Pivot"></a>
Create a pivot table for the `population` and `suicides_no` columns grouped by the age range in the `age` column by using the [`pivot_table` function](https://pandas.pydata.org/docs/reference/api/pandas.pivot_table.html#pandas.pivot_table). 
* Specify the `index` parameter as `age` so that the indexes (row labels) of the returned object correspond to the age ranges.
* Specify the `aggfunc` parameter as `np.sum` (or `"sum"`) to compute aggregated values by taking the sum. The default option is `"mean"`. Note that you do not have to use the `np.sum` option because you divide the `suicides_no` column by the `population` column to compute the rates. However, this example uses the `np.sum` option to display the total number of suicides.

```python
df_age = pd.pivot_table(df, values=["suicides_no","population"], index="age", aggfunc=np.sum)
df_age
```

Reorder the rows and columns by using the [`reindex` method](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.reindex.html#pandas.DataFrame.reindex) so that the indexes are from the younger to the older, and the columns are in alphabetical order.

```python
idx = df_age.index
df_age = df_age.reindex(idx[[3,0,1,2,4,5]], columns=["suicides_no","population"])
df_age
```

When you create a pivot table, if you specify the `columns` parameters instaed of `index`, the column labels of the returned object correspond to the age ranges.

```python
pd.pivot_table(df, values=["suicides_no","population"], columns="age", aggfunc=np.sum)
```

Add a column containing the suicide rates (number of suicides per 100k population) to `df_age`.

```python
df_age["suicides/100k pop"] = df_age["suicides_no"]/df_age["population"]*100000
df_age
```

Alternatively, you can create the table with aggregated values by using the [`groupby` method](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html#pandas.DataFrame.groupby).

```python
df_age2 = df[["suicides_no","population","age"]].groupby(by="age").sum()
df_age2 = df_age2.reindex(idx[[3,0,1,2,4,5]])
df_age2["suicides/100k pop"] = df_age2["suicides_no"]/df_age2["population"]*100000
df_age2
```

## Create Bar Plot <a name="Bar"></a>

Create bar plots for the suicide rates and the total number of suicides by using the [`bar` function](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.bar.html#matplotlib.axes.Axes.bar). 

```python
# Create a figure with two subplots, and set up a wide figure. 
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=plt.figaspect(0.25))

# Create a list of index values after removing the phrase " years". 
idx = df_age.index
idx_short = [text.replace(" years", "") for text in list(idx)]

# Create a bar graph for the suicide rates.
ax1.bar(idx_short, df_age["suicides/100k pop"])

# Specify the labels for the x-axis and y-axis.
ax1.set_xlabel("Age Range")
ax1.set_ylabel("Number of Suicides per 100k people")

# Create a bar graph for the number of suicides.
ax2.bar(idx_short, df_age["suicides_no"])
ax2.set_xlabel("Age Range")
ax2.set_ylabel("Number of Suicides")

# Display the figure.
plt.show()
```

Note that the age range 35-54 has the maximum value for the total number, but the age range 75+ has the maximum rate.

# Rate vs. Generation <a name="RateGeneration"></a>
Examine whether the suicide rate in each age group varies depending on generations.

Create a pivot table for the `population` and `suicides_no` columns grouped by the `age` and `generation` columns.

```python
df_AG = pd.pivot_table(df,values=['population','suicides_no'],index=['age','generation'],aggfunc='sum')
df_AG["suicides/100k pop"] = df_AG["suicides_no"]/df_AG["population"]*100000
df_AG
```

Pivot the outer-most index labels (age) to column labels for the rate values by using the [`unstack` method](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.unstack.html#pandas.DataFrame.unstack).

```python
df_AG_rate = df_AG["suicides/100k pop"].unstack()
df_AG_rate
```

Reorder the table so that the indexes are from the younger to the older and the columns are in chronological order.

```python
idx = df_AG_rate.index
c = df_AG_rate.columns
df_AG_rate = df_AG_rate.reindex(idx[[3,0,1,2,4,5]], columns= c[[3,4,2,0,5,1]])
df_AG_rate
```

More recent generations shows lower rates in general. However, Generation X had lower rates than more recent generations in the age range 5-14, and Bommers had the highest rates in the age range 35-54.

# Rate vs. Year <a name="RateYear"></a>

The table grouped by age and generation shows some trend across years. Examine the yearly trend by creating a pivot table group by the `year` column and creating a line plot by using the [`plot` function](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.plot.html).

```python
df_year = pd.pivot_table(df, values=["population","suicides_no"], index="year", aggfunc="sum")
df_year["suicides/100k pop"] = df_year["suicides_no"]/df_year["population"]*100000
df_year
```

```python
fig, ax = plt.subplots(figsize=plt.figaspect(0.5))

ax.plot(df_year.index, df_year["suicides/100k pop"], linewidth=2.0)

ax.set_xlabel("Year")
ax.set_ylabel("Number of Suicides per 100k people")

plt.grid(visible=True)
plt.show()
```

# Rate vs. Gender <a name="RateGender"></a>

Examine whether the yearly trend differs for each gender.

```python
df_GY = pd.pivot_table(df, values=["suicides_no","population"], index=["sex","year"], aggfunc="sum")
df_GY["suicides/100k pop"] = df_GY["suicides_no"]/df_GY["population"]*100000
df_GY_unstacked = df_GY["suicides/100k pop"].unstack() 
df_GY_unstacked
```

```python
fig, ax = plt.subplots(figsize=plt.figaspect(0.5))

ax.plot(df_year.index, df_year["suicides/100k pop"], linewidth=2.0, label="Female & Male")
ax.plot(df_GY_unstacked.loc["female"].index, df_GY_unstacked.loc["female"], label="Female")
ax.plot(df_GY_unstacked.loc["male"].index, df_GY_unstacked.loc["male"], label="Male")

ax.set_xlabel("Year")
ax.set_ylabel("Number of Suicides per 100k people")

plt.grid(visible=True,which="both")
plt.legend()
plt.show()
```

The yearly trend varies depending on gender. The rates for male has a peak in 1995 while the rates for female show a decreasing trend since 1985. 
In average, the rate for female is less than the rate for male. Confirm the average rates by creating a pivot table grouped by gender.

```python
df_gender = pd.pivot_table(df, values=["suicides_no","population"], index="sex", aggfunc="sum")
df_gender = df_gender.reindex(columns=["suicides_no","population"])
df_gender["suicides/100k pop"] = df_gender["suicides_no"]/df_gender["population"]*100000
df_gender
```

# Rate vs. Country <a name="RateCountry"></a>
Investigate the rates for each county.

Before creating an average table, count the number of rows for each county and year using the `country-year` column. Specify the `aggfunc` parameter of the `pivot_table` function as `"count"`.

```python
df_country_count = pd.pivot_table(df, values=["country-year"], index=["country","year"], aggfunc="count")
df_country_count
```

Some countries have 10 rows in `df` for a certain year. Display the corresponding countries and years.

Because the number of rows varies depending on year for some countries, you can not create an average pivot table simply in one function call. First, create a pivot table grouped by the `country` and `country-year` columns, and then create a pivot table again by grouping `country`. If you do not specify the `aggfunc` parameter, the function computes grouped averages.

```python
df_country_temp = pd.pivot_table(df,values=["HDI for year","gdp_for_year ($)","gdp_per_capita ($)"], index=["country","country-year"])
df_country_temp
```

```python
df_country1 = pd.pivot_table(df_country_temp,values=["HDI for year","gdp_for_year ($)","gdp_per_capita ($)"], index="country")
df_country1
```

Alternatively, you can remove duplicate rows by using the [`drop_duplicates` function](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop_duplicates.html#pandas.DataFrame.drop_duplicates). The values in the three columns (`HDI for year`,`gdp_for_year ($)`, and `gdp_per_capita ($)`) do not vary for a common country-year value. Therefore, you can drop the duplicates based on the `country-year` column.

```python
df_country1_alternative = df.drop_duplicates(subset="country-year")
df_country1_alternative = pd.pivot_table(df_country1_alternative,values=["HDI for year","gdp_for_year ($)","gdp_per_capita ($)"], index="country")
df_country1_alternative
```

Create another pivot table for the rate grouped by `country`. 

```python
df_country2 = pd.pivot_table(df,values=["suicides_no","population"],index="country", aggfunc="sum")
df_country2 = df_country2.reindex(columns=["suicides_no","population"])
df_country2["suicides/100k pop"] = df_country2["suicides_no"]/df_country2["population"]*100000
df_country2
```

Join `df_country1` and `df_country2` by using the [`join` function](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.join.html#pandas.DataFrame.join), and sort tha table by the rates using the [`sort_values` function](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_values.html#pandas.DataFrame.sort_values).

```python
df_country = df_country1.join(df_country2, lsuffix="_caller", rsuffix="_other")
df_country.sort_values(by="suicides/100k pop", inplace=True)
df_country
```

Now you can use the values in `df_country` to examine country-specific characteristics.

Create a horizontal bar plot displaying the rate for each country by using the [`barh` function](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.barh.html#matplotlib.axes.Axes.barh)

```python
fig, ax = plt.subplots(figsize=(8, 20))
ax.barh(df_country.index, df_country["suicides/100k pop"])
ax.set_xlabel("Number of Suicides per 100k people")
ax.set_ylabel("Country")
plt.grid(visible=True,axis="x")
plt.show()
```

Create scatter plots of the rates and other metrics.

```python
fig, ax = plt.subplots()
ax.scatter(df_country["HDI for year"],df_country["suicides/100k pop"])
ax.set_xlabel("HDI")
ax.set_ylabel("Number of Suicides per 100k people")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(df_country["population"],df_country["suicides/100k pop"])
ax.set_xlabel("population")
ax.set_ylabel("Number of Suicides per 100k people")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(df_country["gdp_for_year ($)"],df_country["suicides/100k pop"])
ax.set_xlabel("GDP")
ax.set_ylabel("Number of Suicides per 100k people")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(df_country["gdp_per_capita ($)"],df_country["suicides/100k pop"])
ax.set_xlabel("GDP per Capita")
ax.set_ylabel("Number of Suicides per 100k people")
plt.show()
```

The counties of which GDP per Capita is greater than \$40000 has the rate lower than 20. The other three metrics (HDI, population, and GDP) do not show strong correlations with the rates. 

Examine whether each country shows a different yearly trend.

```python
df_CY = pd.pivot_table(df, values=["suicides_no","population"], index=["country","year"], aggfunc="sum")
df_CY["suicides/100k pop"] = df_CY["suicides_no"]/df_CY["population"]*100000
df_CY_unstacked = df_CY["suicides/100k pop"].unstack() 
df_CY_unstacked

fig, ax = plt.subplots(figsize=(20,10))

year_index = df_CY_unstacked.iloc[0].index
for i in range(len(df_CY_unstacked.index)):
    ax.plot(year_index, df_CY_unstacked.iloc[i])

ax.set_xlabel("Year")
ax.set_ylabel("Number of Suicides per 100k people")
    
plt.grid(visible=True,which="both")
plt.show()
```

It is not easy to distinguish each line, but you can see that some countries have different trends.

Create a pivot plot containing the yearly trend for each country. Also, create the yearly trend table for female and male and for each country.

```python
df_GCY = pd.pivot_table(df,values=["population","suicides_no"],index=["sex","country","year"],aggfunc="sum")
df_GCY["suicides/100k pop"] = df_GCY["suicides_no"]/df_GCY["population"]*100000
df_GCY
```

```python
df_GCY_unstacked = df_GCY["suicides/100k pop"].unstack()
# df_GCY_unstacked
```

```python
df_YC_female = df_GCY_unstacked.loc["female"]
df_YC_male = df_GCY_unstacked.loc["male"]
```

Define a function that creates the yearly trend plots for a selected country.

```python
def plot_ratevscountry(countryIdx):
    fig, ax = plt.subplots()
    
    year_index = df_CY_unstacked.iloc[0].index
    ax.plot(year_index, df_CY_unstacked.iloc[countryIdx], label="female & male")
    ax.plot(year_index, df_YC_female.iloc[countryIdx], label="femele")
    ax.plot(year_index, df_YC_male.iloc[countryIdx], label="male")
    
    ax.set_xlabel("Year")
    ax.set_ylabel("Number of Suicides per 100k people")

    plt.title(df_CY_unstacked.index[countryIdx])
    plt.grid(visible=True,which="both")
    plt.legend()
    plt.show()
```

Display the list of countries so that you can use the index when you call the `plot_ratevscountry` function.

Plot the trend lines for Republic of Korea.

```python
plot_ratevscountry(73)
```

This country has a different trend compared to the average trend. The rates are higher for male, however the yearly trends are similar for female and male.
